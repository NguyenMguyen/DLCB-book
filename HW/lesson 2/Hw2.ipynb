# Bài 1

```python
def grad(x):
    return 2*x


def loss(x):
    return x**2


def gd(x, lr=0.4):
    x_hist = [x]
    loss_hist = [loss(x)]
    
    for i in range(100):
        x = x - lr*grad(x)
        if abs(grad(x)) < 1e-9:
            break
        x_hist.append(x)
        loss_hist.append(loss(x))

    return (x_hist, loss_hist)
```

### Learning rate vừa đủ = 0.4

```python
(x_hist, loss_hist) = gd(5, 0.4)

plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

### Learning rate vừa đủ = 0.5
<br/>
Trường hợp đặc biệt: hội tụ ngay lập tức

```python
(x_hist, loss_hist) = gd(5, 0.5)

plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

### Learning rate nhỏ = 0.05

```python
(x_hist, loss_hist) = gd(5, 0.05)

plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

# Bài 3

### a) Lúc em làm bài này chưa có dữ liệu trên drive nên em tự sinh dữ liệu rồi làm.

```python
w = np.array([[2], [1]])
```

```python
X = np.random.randint(50, size=100) + np.random.randn(100)
X = X.reshape(-1, 1)
ones = np.ones_like(X)
X = np.concatenate((X, ones), axis=1)
y = X.dot(w) + np.random.randn(100).reshape(-1, 1)
y = y.reshape(-1, 1)
N = X.shape[0]
```

```python
plt.scatter(X[:, 0], y)
plt.show()
```

### Triển khai model

```python
def loss(w):
    return 0.5/N * (np.linalg.norm(y-X.dot(w))**2)


def grad(w):
    return 1/N * X.T.dot(X.dot(w)-y)


def gd_linear(w, lr):
    loss_hist = [loss(w)]
    w_hist = [w]
    for i in range(10000):
        w_new = w_hist[-1] - lr*grad(w_hist[-1])
        if np.linalg.norm(grad(w_new))/len(w_new) < 1e-3:
            break
        loss_hist.append(loss(w_new))
        w_hist.append(w_new)

    return (loss_hist, w_hist)
```

```python
w_init = np.random.randn(2,1)
(loss_hist, w_hist) = gd_linear(w_init, 0.001)
```

```python
plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

```python
# Draw the fitting line
x0 = np.linspace(1, 50, 2, endpoint=True).reshape(-1,1)
x0 = np.concatenate((x0, np.ones_like(x0)), axis=1)
y0 = x0.dot(w_hist[-1])
plt.tick_params(axis='both', which='major', labelsize=13)
plt.scatter(X[:, 0], y)
plt.plot(x0[:, 0], y0, 'y', linewidth=4)
plt.show()
```

### Learning rate lớn

```python
(loss_hist, w_hist) = gd_linear(w_init, 0.005)
plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

### Learning rate nhỏ

```python
(loss_hist, w_hist) = gd_linear(w_init, 0.000001)
plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

### b) Dữ liệu linear.csv

```python
d = pd.read_csv('data_linear.csv').values
x_raw = d[:, 0].reshape(-1, 1)
y_raw = d[:, 1].reshape(-1, 1)
N = x_raw.shape[0]
```

```python
plt.scatter(x_raw, y_raw)
plt.show()
```

#### Chuẩn hoá dữ liệu

```python
max_x = np.max(x_raw)
min_x = np.min(x_raw)
max_y = np.max(y_raw)
min_y = np.min(y_raw)
x = (x_raw - min_x) / (max_x - min_x)
y = (y_raw - min_y) / (max_y - min_y)
```

```python
Xbar = np.concatenate((np.ones_like(x), x), axis=1)
```

```python
def loss(w):
    return 0.5/N * (np.linalg.norm(Xbar.dot(w)-y)**2)


def grad(w):
    return 1/N * Xbar.T.dot(Xbar.dot(w)-y)


def gd_linear(w, lr):
    loss_hist = [loss(w)]
    
    for i in range(1000):
        w -= lr*grad(w)
        if np.linalg.norm(grad(w))/len(w) < 1e-3:
            break
        loss_hist.append(loss(w))

    return (w, loss_hist)
```

```python
w_init = np.random.randn(2,1)
(w, loss_hist) = gd_linear(w_init, 1)
print(w)
print(loss_hist[-1])
```

```python
plt.plot(loss_hist)
plt.xlabel('number of epoches', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

```python
# Draw the fitting line
x0_raw = np.linspace(30, 100, 2, endpoint=True).reshape(-1,1)
x0 = (x0_raw - min_x) / (max_x - min_x)
x0 = np.concatenate((np.ones_like(x0), x0), axis=1)
y0 = x0.dot(w)
y_true = y0 * (max_y - min_y) + min_y

plt.tick_params(axis='both', which='major', labelsize=13)
plt.scatter(x_raw, y_raw)
plt.plot(x0_raw, y_true, 'y', linewidth=4)
plt.show()
```

# Bài 4

```python
d = pd.read_csv('data_square.csv').values
x_raw = d[:, 0].reshape(-1, 1)
y_raw = d[:, 1].reshape(-1, 1)
N = x_raw.shape[0]
```

### Biểu diễn dữ liệu

```python
plt.scatter(x_raw, y)
plt.show()
```

Xây dựng model
<br>
Phương trình model có dạng y = a\*x^2 + b\*x +c
<br>
Đầu tiên, chuẩn hoá x và y về khoảng [0,1]

```python
a, b = 0, 1
x = a + (x_raw - np.min(x_raw)) * (b-a) / (np.max(x_raw) - np.min(x_raw))
```

```python
Xbar = np.concatenate((np.ones_like(x), x, x**2), axis=1)
```

```python
y = (y_raw - np.min(y_raw)) / (np.max(y_raw) - np.min(y_raw))
```

### Các hàm chính

```python
def loss(w):
    return 0.5/N * (np.linalg.norm(Xbar.dot(w)-y)**2)


def grad(w):
    return 1/N * Xbar.T.dot(Xbar.dot(w)-y)


def gd_linear(w, lr):
    loss_hist = [loss(w)]
    
    for i in range(1000):
        w -= lr*grad(w)
        if np.linalg.norm(grad(w))/len(w) < 1e-5:
            break
        loss_hist.append(loss(w))

    return (w, loss_hist)
```

```python
w_init = np.array([1.,1.,1.]).reshape(-1, 1)
(w, loss_hist) = gd_linear(w_init, 1.4)
print(w)
print(len(loss_hist))
print(loss_hist[-1])
```

### Biểu diễn đồ thị tìm được

```python
x0_raw = np.linspace(30, 100, 500, endpoint=True).reshape(-1,1)
a, b = 0, 1
x0 = a + (x0_raw - np.min(x0_raw)) * (b-a) / (np.max(x0_raw) - np.min(x0_raw))

X0 = np.concatenate((np.ones_like(x0), x0, x0**2), axis=1)
y0 = X0.dot(w)
y0_true = y0 * (np.max(y_raw) - np.min(y_raw)) + np.min(y_raw)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.plot(x0_raw, y0_true, 'y', linewidth=4)
plt.show()
```
